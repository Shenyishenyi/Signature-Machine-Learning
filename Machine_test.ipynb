```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams["text.usetex"]=True
```

```python
gap=np.arange(0,1,0.01)
x=np.cos(10*gap)
y=np.sin(3*gap)
plt.plot(x,y)
```

```python

x_reparmetrized=np.cos(10*gap)
y_reparametrized=np.sin(3*gap)
plt.figure(figsize=(7,7))
plt.plot(x_reparmetrized,y_reparametrized,color="r",linestyle="-",label=r"reparametrized path $\gamma _t$")
plt.plot(x,y,color="b",linestyle="--",label="orgianl path")
plt.legend(loc="lower left")
```
