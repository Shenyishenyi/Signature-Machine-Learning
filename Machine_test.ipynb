```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
gap=np.arange(0,1,0.01)
x=np.cos(10*gap)
y=np.sin(3*gap)
plt.plot(x,y)
```

```python

x_reparmetrized=np.cos(10*gap)
y_reparametrized=np.sin(3*gap)
plt.plot(x_reparmetrized,y_reparametrized,color="r",linestyle="-")
plt.plot(x,y,color="b",linestyle="--")
```
